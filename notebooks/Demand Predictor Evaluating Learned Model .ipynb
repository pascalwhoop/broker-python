```python
def plt_10_times(model, seqs):
    import random, pandas
    for i in range(10):
        cust = random.choice(seqs)
        batch = random.choice(cust)
        X = batch[0][0]
        X_FLAT = X.flatten()
        print(X.shape)
        Y = batch[1][0]
        PRED = model.predict(np.array([X])).flatten()
        real = np.concatenate((X.flatten(),Y))
        print(real.shape)
        pred = np.full(X_FLAT.shape, np.nan)
        pred = np.concatenate((pred,PRED))
        df = pandas.DataFrame(np.array([real,pred]).transpose())
        df.plot(figsize=(12,8))

```

```python
plt_10_times(dl, seqs)
```
